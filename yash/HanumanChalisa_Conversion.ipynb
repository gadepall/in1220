```
text=open('hanuman_chalisa.txt','r')
```

```
text=text.read()
```

```
text
```

```
##chaupai starts after the key word चौपाई
all_chaupai=text.split('चौपाई')[1]
```

```
all_chaupai
```

```
individual_chaupai=all_chaupai.split('\n\n')
individual_chaupai
```

Now let's start with our unicode conversion !

```
unicode_all_chaupai=[]
```

```
#for each chaupai we will create a unicode array and then append this array to our unicode_all_array
for chaupai in individual_chaupai:
  arr=[]
  print(chaupai,"\n\n")
  for word in chaupai.split():
    print(word, word.encode())
    arr.append(word.encode())

  print()
  unicode_all_chaupai.append(arr)
```
